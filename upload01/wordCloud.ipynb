```python
import wordcloud #导入词云库
import numpy as np
import matplotlib.pyplot as plt
import PIL
import jieba
import re
from wordcloud import WordCloud,STOPWORDS, ImageColorGenerator
```

```python
with open(r'0824.txt') as f:
    text1 = f.readlines()
#导入图片
image1 = PIL.Image.open(r'D:\txt\map_weibo.jpg')
MASK = np.array(image1)
stop_words = ["信息", "来源","新浪","人民","日报","闪电","新闻","人民日报"]
WC = wordcloud.WordCloud(font_path = 'C:\\Windows\\Fonts\\msyh.ttc',
                         max_words=2000,
                         mask = MASK,
                         height= 600,
                         width=600,
                         background_color='white',
                         stopwords = stop_words
                        )
                         
#                          repeat=False,
#                          mode='RGBA') #设置词云图对象属性
st1 = re.sub('[，。、“”‘ ’]','',str(text1)) #使用正则表达式将符号替换掉。
conten = ' '.join(jieba.lcut(st1)) #此处分词之间要有空格隔开，联想到英文书写方式，每个单词之间都有一个空格。
con = WC.generate(conten)
plt.imshow(con)
plt.axis("off")
plt.savefig("D:\\txt\\jpg\\weibo_words",dpi=1000)
# wc.to_file(path+r'\beautifulcloud.png') 
```
